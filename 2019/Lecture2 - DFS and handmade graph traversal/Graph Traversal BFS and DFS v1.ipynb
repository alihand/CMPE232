```python
class node():
    def __init__(self, name):
        self.name = name
        self.neighbors = []
        self.visited = False
        self.prev = None
        
class BreadthFirstSearch():
    def __init__(self):
        self.queue = []
    
    def bfs(self, s):
        self.queue.append(s)
        s.visited = True
        
        while len(self.queue) != 0:
            v = self.queue.pop(0)
            print("%s " % v.name);
            
            for neighbor in v.neighbors:
                if not neighbor.visited:
                    self.queue.append(neighbor)
                    neighbor.visited = True
                    neighbor.prev = v
        
```

```python
## Recursive DFS approach with implicit stack

class node():
    def __init__(self, name):
        self.name = name
        self.neighbors = []
        self.visited = False
        self.prev = None
        
class RecursiveDepthFirstSearch():    
    def dfs(self, s):
        s.visited = True
        print("%s " % s.name);
        for neighbor in s.neighbors:
            if not neighbor.visited:
                self.dfs(neighbor)
```

```python
nodeA = node("A");
nodeB = node("B");
nodeC = node("C");
nodeD = node("D");
nodeE = node("E");
nodeF = node("F");

nodeA.neighbors.append(nodeB);
nodeA.neighbors.append(nodeC);
nodeA.neighbors.append(nodeE);

nodeB.neighbors.append(nodeA);
nodeB.neighbors.append(nodeC);

nodeC.neighbors.append(nodeA);
nodeC.neighbors.append(nodeD);
nodeC.neighbors.append(nodeB);

nodeD.neighbors.append(nodeE);
nodeD.neighbors.append(nodeF);
nodeD.neighbors.append(nodeC);

nodeE.neighbors.append(nodeD);

nodeF.neighbors.append(nodeD);

dfs = RecursiveDepthFirstSearch();
dfs.dfs(nodeA);
```
